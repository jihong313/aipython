```python
height1 = float(input("학생 1의 키를 cm로 입력하시오."))
height2 = float(input("학생 2의 키를 cm로 입력하시오."))

if height1 == height2:
  print("두 학생의 키는 같다.")

if height1 > height2:
   print("첫 번째 학생이 더 크다.")

if height1 < height2:
  print("두 번째 학생이 더 크다.")
```

```python
height1 = float(input("학생 1의 키를 cm로 입력하시오."))
height2 = float(input("학생 2의 키를 cm로 입력하시오."))

if height1 == height2:
  print("두 학생의 키는 같다.")

elif height1 > height2:
 print("첫 번째 학생이 더 크다.")
 print(f"두사람의 키 차이는:{height1-height2}")

else:
  print("두 번째 학생이 더 크다.")
  print(f"두사람의 키 차이는:{height2-height1}")


```

```python
#점수를 입력받아 90점 이상 A 80점 이상 B 70점 이상 C 나머지 D로 출력하시오
score = int(input("점수를 입력하시오."))
if score >= 90:
  print("A")

elif score >= 80:
  print("B")

elif score >= 70:
  print("C")

else:
  print("D")
```

```python

import random

dice1=random.randint(1,6)
dice2=random.randint(1,6)

print(f'주사위1:{dice1}')
print(f'주사위2:{dice2}')

if dice1 > dice2:
  print("주사위1이 크다.")

elif dice1 < dice2:
  print("주사위2가 크다.")

else:
  print("주사위1과 주사위2가 같다.")
```
