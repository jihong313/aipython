```python
# 사용자에게 "분" 단위로 시간을 입력받아 "ㅁ시 ㅁㅁ분 입니다." 형태로 출력

total_minutes = int(input("시간을 분 단위로 입력하세요: "))

hour = total_minutes // 60      # 시간(몫)
minute = total_minutes % 60     # 분(나머지)

print(f"{hour}시 {minute:02d}분 입니다.")
```

```python
# 사용자에게 숫자를 입력받아 짝수/홀수를 판별하는 프로그램

# 1) 사용자에게 숫자를 입력받는다 (입력은 문자열이므로 int로 변환)
num = int(input("숫자를 입력하세요: "))

# 2) 나머지 연산자(%)를 사용한다
#    num % 2 의 결과가 0이면 2로 나누어 떨어지므로 짝수
if num > 0:
    print("양수입니다.")
    else:
    print("음수입니다.")
```

```python
price_coffee=2000
price_latte=3000
price_tea=2500
price_juice=2800
order_c=int(input('커피 주문 수:'))

order_l=int(input('라떼 주문 수:'))

order_t=int(input('녹차 주문 수:'))

order_j=int(input('주스 주문 수:'))

order_cost = price_coffee*order_c+price_latte*order_l+price_tea*order_t+price_juice*order_j

print(f"주문 금액은 {order_cost} 원 입니다.")
```
