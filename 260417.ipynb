```python
sum=0
while True:
 for i,a in menu.items():
  print(f"{i}는 {a}원입니다.")
 print("="*10)
 i = input("메뉴를 입력하거나 종료를 입력하세요: ")#.strip()

 if i == "종료":
    break
 if i not in menu:
    print("해당 메뉴는 없습니다. 다시 입력하세요.")
    continue
 q=int(input("수량을 입력하세요."))
 #숫자로 입력하지 않거나 0으로 입력할 경우 예외처리
 price= menu[i]*q
 sum+=price
 print("="*10)
 print(f"현재 누적금액:{price}")
 print("="*10)
print(f"총 금액은 {sum}입니다.")

```
