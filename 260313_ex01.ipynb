```python
# 사칙 연산 프로그램 (변수 + 연산자 + 출력문만 사용)

a = 12          # 첫 번째 숫자를 a에 저장
b = 3           # 두 번째 숫자를 b에 저장

add = a + b     # 더하기 결과
sub = a - b     # 빼기 결과
mul = a * b     # 곱하기 결과
div = a / b     # 나누기 결과 (파이썬은 / 하면 소수점이 나올 수 있어요)

print("a =", a)                 # a 출력
print("b =", b)                 # b 출력

print("a + b =", add)           # 더하기 결과 출력
print("a - b =", sub)           # 빼기 결과 출력
print("a * b =", mul)           # 곱하기 결과 출력
print("a / b =", div)           # 나누기 결과 출력
```

```python
# 사칙 연산 프로그램 (입력 + 변수 + 연산자 + 출력문)

a = float(input("첫 번째 숫자 a를 입력하세요: "))  # 입력받은 값을 숫자로 변환해 a에 저장
b = float(input("두 번째 숫자 b를 입력하세요: "))  # 입력받은 값을 숫자로 변환해 b에 저장

add = a + b     # 더하기 결과
sub = a - b     # 빼기 결과
mul = a * b     # 곱하기 결과
div = a / b     # 나누기 결과

print("a =", a)                 # a 출력
print("b =", b)                 # b 출력

print("a + b =", add)           # 더하기 결과 출력
print("a - b =", sub)           # 빼기 결과 출력
print("a * b =", mul)           # 곱하기 결과 출력
print("a / b =", div)           # 나누기 결과 출력
```
