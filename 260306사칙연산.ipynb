```python
# 두 개의 숫자를 입력받아 사칙연산 결과를 출력하는 코드

a = float(input("첫 번째 숫자를 입력하세요: "))
b = float(input("두 번째 숫자를 입력하세요: "))

print(f"덧셈: {a} + {b} = {a + b}")
print(f"뺄셈: {a} - {b} = {a - b}")
print(f"곱셈: {a} * {b} = {a * b}")

if b != 0:
    print(f"나눗셈: {a} / {b} = {a / b}")
else:
    print("나눗셈: 0으로 나눌 수 없습니다.")
```
