```python
import random
answer=random.randint(1,100)
```

```python
answer
```

```python
user=int(input("추측값을 입력하시오."))
```

```python
count=0
while user != answer:
  if user > answer:
    print("더 작은 수을 입력하세요.")
    count+=1
  else:
    print("더 큰 수를 입력하세요.")
    count+=1
  user=int(input("추측값을 입력하시오."))
print("총 시도 횟수는",count)
```

```python
num = int(input("정수를 입력하시오."))
for i in range (1, num+1):
  if num % i == 0:
    print(i)
```
